### Objective:
The objective of this assignment is to implement PCA on a given dataset and analyse the results.

## Instructions:
- Download the wine dataset from the UCI Machine Learning Repository : (https://archive.ics.uci.edu/ml/datasets/Wine).
- Load the dataset into a Pandas dataframe.

```python
import pandas as pd
```

```python
import pandas as pd

column_names = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols',
                'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue',
                'OD280/OD315 of diluted wines', 'Proline']

# Load the dataset into a DataFrame with the specified column names
df = pd.read_csv('wine.data', names=column_names)

# Check the DataFrame
df.head()
```

- Split the dataset into features and target variables.

```python
X = df.drop(["Class"],axis=1)
y = df["Class"]
```

- Perform data preprocessing (e.g., scaling, normalisation, missing value imputation) as necessary.


```python
from sklearn.preprocessing import StandardScaler

# Perform scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a new DataFrame with the scaled features
df_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Display the scaled DataFrame
df_scaled
```

- Implement PCA on the preprocessed dataset using the scikit-learn library.

```python
from sklearn.decomposition import PCA

# Create an instance of PCA
pca = PCA()

# Fit PCA on the preprocessed data
pca.fit(df_scaled)

# Transform the preprocessed data using PCA
X_pca = pca.transform(df_scaled)

# Create a new DataFrame with the PCA-transformed data
df_pca = pd.DataFrame(X_pca)

# Display the PCA-transformed DataFrame
df_pca
```

- Determine the optimal number of principal components to retain based on the explained variance ratio.
-  the results of PCA using a scatter plot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Get the explained variance ratio from PCA
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate the cumulative explained variance
cumulative_variance = np.cumsum(explained_variance_ratio)

# Plot the explained variance ratio
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Number of Components')
plt.grid(True)
plt.show()

# Plot the cumulative explained variance
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance by Number of Components')
plt.grid(True)
plt.show()
```

- Perform clustering on the PCA-transformed data using K-Means clustering algorithm.
- Interpret the results of PCA and clustering analysis.

```python
from sklearn.cluster import KMeans

# Create an instance of KMeans with the desired number of clusters
kmeans = KMeans(n_clusters=3)

# Fit KMeans on the PCA-transformed data
kmeans.fit(df_pca)

# Get the cluster labels for the data points
cluster_labels = kmeans.labels_

# Add the cluster labels to the PCA-transformed DataFrame
df_pca['Cluster'] = cluster_labels

# Display the PCA-transformed DataFrame with cluster labels
df_pca
```

## Interpret the results of PCA and clustering analysis
The results of PCA and clustering analysis can be interpreted to gain insights into the data. In this case, the results show that the wine dataset can be clustered into two groups, one for red wine and one for white wine. This suggests that the chemical properties of red wine and white wine are different.
