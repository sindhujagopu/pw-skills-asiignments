```python
Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

When creating a custom exception in Python, it is important to inherit from the built-in Exception class or one of its subclasses. This is because Exception provides a set of attributes and methods that are necessary for properly handling and reporting errors in a program.
Here are a few reasons why we should use the Exception class as the base class for custom exceptions:
Inheritance: The Exception class provides a well-defined hierarchy for handling exceptions. By inheriting from Exception, our custom exception becomes a part of this hierarchy and can be caught and handled along with other built-in exceptions.
Error Reporting: The Exception class provides attributes like args and message that can be used to provide more detailed information about the error that occurred. By using these attributes in our custom exception, we can provide better error reporting for our users.
Standardization: By using the Exception class as the base class for our custom exception, we are following a well-established convention in Python. This makes our code more understandable and maintainable for other developers who might be working on the same project.
Overall, using the Exception class as the base class for custom exceptions ensures that our exceptions are compatible with the rest of the Python exception hierarchy and provides the necessary attributes and methods for handling and reporting errors.
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is a built-in exception class in Python that is raised when a module cannot be imported. This error typically occurs when the Python interpreter is unable to find the module specified in the import statement, or when the module is found but cannot be loaded for some reason.
For example, consider the following code:

import non_existent_module   # ImportError: No module named 'non_existent_module'

In this case, the ImportError is raised because there is no module named non_existent_module that can be imported.
ModuleNotFoundError is a subclass of ImportError that was introduced in Python 3.6 to provide a more informative error message when a module cannot be found. Prior to Python 3.6, the ImportError exception was used for both cases where a module could not be found and for other types of import errors.
For example, consider the following code:
    
import non_existent_module   # ModuleNotFoundError: No module named 'non_existent_module'

In this case, the ModuleNotFoundError is raised instead of ImportError because the module non_existent_module could not be found.
Overall, ImportError and ModuleNotFoundError are both important exceptions in Python that indicate that a module could not be imported for some reason. The difference between them is that ModuleNotFoundError provides a more specific error message when the module cannot be found, while ImportError is a more general error that can be raised for other types of import errors as well.
```

```python
Q6. List down some best practices for exception handling in python.

Only catch exceptions that you can handle: It's important to catch only the exceptions that you can handle in a meaningful way. Catching all exceptions can make it difficult to debug issues and can also hide underlying problems.

Use specific exception classes: Catching specific exception classes instead of using a general catch-all statement can help to make the code more readable and also provide more specific information about the error.

Handle exceptions at the appropriate level: Exceptions should be handled at the appropriate level of the code. For example, if an exception occurs in a function, it should be caught and handled within that function, rather than being propagated up to the higher level of the code.

Use finally blocks for cleanup: The finally block is always executed, regardless of whether an exception is raised or not. This makes it a good place to put code that needs to be executed for cleanup purposes, such as closing files or database connections.

Avoid using bare except clauses: Using a bare except clause to catch all exceptions can make it difficult to debug issues and can also hide underlying problems. It's better to catch specific exceptions or use a general catch-all clause at the appropriate level of the code.

Don't suppress exceptions: Suppressing exceptions by catching them and not doing anything with them can lead to unexpected behavior and make it difficult to debug issues. Always log or handle exceptions in a meaningful way.

Use try-except blocks sparingly: Using too many try-except blocks can make the code harder to read and maintain. It's better to write code that handles errors gracefully without the use of try-except blocks, where possible.
```
